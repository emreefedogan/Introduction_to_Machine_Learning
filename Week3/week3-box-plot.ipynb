```python
# **Box Plot**

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from plotly.offline import iplot
import plotly.graph_objs as go


timesData = pd.read_csv("/kaggle/input/university-rankings-2011/University_Rankings_20220204.csv")
timesData.head()
timesData.info()
x2012 = timesData[timesData.year == 2012] 
international = [float(each) for each in x2012.international] 
                                              
trace0 = go.Box(  
    y=x2012.international, 
    name = 'international score of universities in 2012',
    marker = dict(
        color = 'rgb(12, 12, 140)',
    )
)
trace1 = go.Box(
    y=x2012.research,
    name = 'research of universities in 2012',
    marker = dict(
        color = 'rgb(12, 128, 128)',
    )
)
data = [trace0, trace1]
iplot(data)
```
